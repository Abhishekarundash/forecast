```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("meals_std_alldates.csv")
df = df[df['date']<= '2024-02-11' ]
df = df[df['date']> '2024-01-21' ]
fri = df[["foodcourt","restaurant","restaurantmenuitem"]].drop_duplicates().reset_index(drop=True)
```

```python
# df1 = df[["date","foodcourtid","restaurant","menuitemid","FCRI_total_count"]].dropna()
for index, row in fri.iterrows():
    fc = row["foodcourt"]
    rest = row["restaurant"]
    item = row["restaurantmenuitem"]
    item_ts = df[(df["foodcourt"]==fc)&(df["restaurant"]==rest)&(df["restaurantmenuitem"]==item)]# Filtering item wise
    item_ts1 = item_ts[item_ts["total_count"]>0] # Creating the dataframe with non 0 sales
    fri.at[index,"cov"] = item_ts1["total_count"].std()/item_ts1["total_count"].mean() # Creating COV
    fri.at[index,"sale_data_points"] = item_ts1.shape[0] # Creating count of non zero sales
    item_ts["is_active"] = np.where(item_ts["total_count"]>0,1,0) 
    original_list = item_ts["is_active"].tolist() # Creating a list of 49 elements
    sublists = [original_list[i:i+7] for i in range(0, len(original_list), 7)] #Breaking the list into small list with 7 sublists
    sublists_as_tuples = [tuple(sublist) for sublist in sublists] #Converting sublist to tuple
    num_unique_sublists = len(set(sublists_as_tuples))# Counting unique number of sublists
    fri.at[index,"#unique_pattern"] = num_unique_sublists
```

```python
fri.to_excel("sku_stability2.xlsx")
```
