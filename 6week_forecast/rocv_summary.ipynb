```python
rocv1 = pd.read_excel('rocv1_out.xlsx')
rocv2 = pd.read_excel('rocv2_out.xlsx')
rocv3 = pd.read_excel('rocv3_out.xlsx')
```

```python
rocv_out = pd.merge(rocv1, rocv2, on=['foodcourt', 'restaurant', 'restaurantmenuitem'], how='left')
rocv_out = pd.merge(rocv_out, rocv3, on=['foodcourt', 'restaurant', 'restaurantmenuitem'], how='left')
rocv_out = rocv_out[['foodcourt', 'restaurant', 'restaurantmenuitem','r1_accuracy','r1_model','r2_accuracy','r2_model','r3_accuracy','r3_model']]
```

```python
def most_frequent_model(row):
    models = row[['r1_model', 'r2_model', 'r3_model']]
    model_counts = models.value_counts()
    if model_counts.iloc[0] == model_counts.max() and len(model_counts) > 1 and model_counts.iloc[1] == model_counts.max():
        return row['r3_model']
    else:
        return model_counts.idxmax()
# Apply the function to each row to create the new column
rocv_out['most_frequent_model'] = rocv_out.apply(most_frequent_model, axis=1)
```
